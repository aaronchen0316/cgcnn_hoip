```
import os
import csv
import pymatgen
import numpy as np
import matplotlib.pyplot as plt
from pymatgen import MPRester, Spin, Element
```

```
material_id = []
formula = []
API_KEY = 'qjFRhic2v7SG60lA2JZZ'
mpr = MPRester(API_KEY)
```

```
with open('ABX3.csv') as inputcsv:
  reader = csv.DictReader(inputcsv)
  num_data = 0
  for line in reader:
    material_id.append(line['Materials Id'])
    formula.append(line['Formula'])
    num_data += 1
print(num_data)
```

```
def get_property(material):

  material = material[0]

  try:
    energy = material['energy']
  except AttributeError:
    energy = 'None'

  try:
    formation_energy = material['formation_energy_per_atom'] * material['nsites']
  except AttributeError:
    formation_energy = 'None'
  
  try:
    band_gap = material['band_gap']
  except AttributeError:
    band_gap = 'None'
  
  return energy, formation_energy, band_gap
```

```
energy, formation_energy, band_gap = [], [], []

for i in range(num_data):
  # print(i, material_id[i])
  # print(mpr.get_data(material_id[i]))
  energy_temp, formation_energy_temp, band_gap_temp = get_property(mpr.get_data(material_id[i]))
  # print(energy_temp, formation_energy_temp, band_gap_temp)
  if energy_temp == 'None':
    print('No Energy for %s, line=%d' %(material_id[i], i+2))
  else:
    energy.append(energy_temp)
  
  if formation_energy_temp == 'None':
    print('No Formation Energy for %s, line=%d' %(material_id[i], i+2))
  else:
    formation_energy.append(formation_energy_temp)
  
  if band_gap_temp == 'None':
    print('No Band Gap for %s, line=%d' %(material_id[i], i+2))
  else:
    band_gap.append(band_gap_temp)
```

```
with open('id_prop.csv', 'w') as targetcsv:
  writer = csv.writer(targetcsv, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
  writer.writerow(['material_id', 'energy', 'formation_energy', 'band_gap'])
  for i in range(num_data):
    writer.writerow([material_id[i], energy[i], formation_energy[i], band_gap[i]])
```
